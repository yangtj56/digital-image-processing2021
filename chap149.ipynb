#Contour Analysis

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
im = cv2.imread('Sym.jpg')
im1 = cv2.imread('Sym2.jpg')
im2 = cv2.imread('Hand.jpg')
im3 = cv2.imread('Pot.jpg')
imgray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
imgray1 = cv2.cvtColor(im1,cv2.COLOR_BGR2GRAY)
imgray2 = cv2.cvtColor(im2,cv2.COLOR_BGR2GRAY)
imgray3 = cv2.cvtColor(im3,cv2.COLOR_BGR2GRAY)
ret,thresh = cv2.threshold(imgray,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
ret1,thresh1 = cv2.threshold(imgray1,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
ret2,thresh2 = cv2.threshold(imgray2,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
ret3,thresh3 = cv2.threshold(imgray3,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
contours1, hierarchy1 = cv2.findContours(thresh1,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
contours2, hierarchy2 = cv2.findContours(thresh2,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
contours3, hierarchy3 = cv2.findContours(thresh3,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
im = cv2.drawContours(im, contours, -1, (255,0,0), 3)
im1 = cv2.drawContours(im1, contours1, -1, (0,255,0), 3)
im2 = cv2.drawContours(im2, contours2, -1, (0,0,255), 3)
im3 = cv2.drawContours(im3, contours3, -1, (255,0,0), 3)
```

```python
plt.figure(figsize=(15,10))
plt.subplot(331),plt.imshow(imgray,cmap = 'gray')
plt.title('img'), plt.xticks([]), plt.yticks([])
plt.subplot(332),plt.imshow(thresh,cmap = 'gray')
plt.title('thresh'), plt.xticks([]), plt.yticks([])
plt.subplot(333),plt.imshow(im,cmap = 'gray')
plt.title('drawContours'), plt.xticks([]), plt.yticks([])
plt.subplot(334),plt.imshow(imgray1,cmap = 'gray')
plt.title('img'), plt.xticks([]), plt.yticks([])
plt.subplot(335),plt.imshow(thresh1,cmap = 'gray')
plt.title('thresh'), plt.xticks([]), plt.yticks([])
plt.subplot(336),plt.imshow(im1,cmap = 'gray')
plt.title('drawContours'), plt.xticks([]), plt.yticks([])
plt.subplot(337),plt.imshow(imgray2,cmap = 'gray')
plt.title('img'), plt.xticks([]), plt.yticks([])
plt.subplot(338),plt.imshow(thresh2,cmap = 'gray')
plt.title('thresh'), plt.xticks([]), plt.yticks([])
plt.subplot(339),plt.imshow(im2,cmap = 'gray')
plt.title('drawContours'), plt.xticks([]), plt.yticks([])
plt.show()
```
