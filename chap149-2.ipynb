#Contours in OpenCV

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
im = cv2.imread('The_Ankh.jpg')
img = im.copy()
imgray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
ret,thresh = cv2.threshold(imgray,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
```

```python
im = cv2.drawContours(im,contours,-1,(255,0,0),3)
```

```python
plt.figure(figsize=(12,12))
plt.subplot(131),plt.imshow(imgray,cmap = 'gray')
plt.title('img'), plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(thresh,cmap = 'gray')
plt.title('thresh'), plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(im,cmap = 'gray')
plt.title('drawContours'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
im = cv2.imread('Jiebao.png')
imgray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
blur = cv2.GaussianBlur(im,(9,9),0)
blur2 = cv2.GaussianBlur(im,(9,9),0)
imgray2 = cv2.cvtColor(blur2,cv2.COLOR_BGR2GRAY)
ret,thresh = cv2.threshold(imgray2,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
img = cv2.drawContours(im, contours, -1, (255,0,0), 3)
```

```python
plt.figure(figsize=(20,20))
plt.subplot(121),plt.imshow(imgray,cmap = 'gray')
plt.title('img'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(im,cmap = 'gray')
plt.title('drawContours'), plt.xticks([]), plt.yticks([])
plt.show()
```

1. Moments

```python
cnt = contours[1] 
M = cv2.moments(cnt)
print(M)
```

```python
cx = int(M['m10']/M['m00'])
cy = int(M['m01']/M['m00'])
```

## (w,h,c) = img.shape

```python
img = np.zeros(im.shape,np.unit8)
img = cv2.circle(img,(cx,cy),3,(255,255,0),2)
img = cv2.drawContours(img,contours[1],-1, (255,0,0), 3)
```

```python
plt.figure(figsize=(20,20))
plt.subplot(121),plt.imshow(imgray,cmap = 'gray')
plt.title('img'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(img,cmap = 'gray')
plt.title('drawContours'), plt.xticks([]), plt.yticks([])
plt.show()
```

#4.Contour Approximation

```python
epsilon = 0.001*cv2.arcLength(cnt,True)
approx = cv2.approxPolyDP(cnt,epsilon,True)
```

```python
imgA = np.zeros(im,shape,np,unit8)
imgA = cv2.drawContours(imgA,approx,-1,(0,255,0),3)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Contour Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(imgA,cmap = 'gray')
plt.title('Approx Contour Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

5. Convex Hull

```python
hull = cv2.convexHull(cnt)
```

```python
imgA = np.zeros(im,shape,np,unit8)
imgA = cv2.drawContours(imgA,[hull],-1,(0,255,0),3)
```
